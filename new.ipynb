```python
#Question: What proportion of Netflix content is original vs. licensed?
```

```python
# Importing essential libraries for data analysis and visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading Netflix dataset into a Pandas DataFrame
netflix_df = pd.read_csv('netflix_titles.csv')
```

```python
# Fill missing type values with "Licensed" 
netflix_df['type'] = netflix_df['type'].fillna('Licensed')

# Remove duplicates
netflix_df.drop_duplicates(inplace=True)
```

```python
# Group by type and count titles 
type_counts = netflix_df.groupby('type')['title'].count()

```

```python
# Printing the index of the 'type_counts' Series
print(type_counts.index)
```

```python
# Checking if there are any entries in the DataFrame where the 'type' is 'Original'
print(netflix_df[netflix_df['type'] == 'Original'].shape[0] > 0)
```

```python
# Calculating the proportion of 'Original' entries in the Netflix DataFrame
if netflix_df[netflix_df['type'] == 'Original'].shape[0] == 0:
    original_prop = 0
else:
    original_prop = type_counts.loc['Original'] / sum(type_counts)
```

```python
# Group by type and count titles 
type_counts = netflix_df.groupby('type')['title'].count()

```

```python
# Displaying the proportion of 'Original' content in the Netflix dataset
if netflix_df[netflix_df['type'] == 'Original'].shape[0] == 0:
    original_prop = 0
else: 
    original_prop = type_counts.loc['Original'] / sum(type_counts)

print(f'Proportion of Original Content: {original_prop:.2%}')
```

```python
# Creating a bar plot to visualize the distribution of content types in the Netflix dataset
ax = netflix_df['type'].value_counts().plot(kind='bar')
ax.set_ylabel('Count')
ax.set_title('Content Distribution by Type')
```

```python
# Extracting and visualizing the top 10 most common genres in the Netflix dataset
top_genres = netflix_df.groupby('listed_in')['title'].count().sort_values(ascending=False).head(10)

ax = top_genres.plot(kind='bar')
ax.set_title('Most Common Genres')
ax.set_ylabel('Number of Titles')
```

```python

# Creating a distribution plot for Netflix ratings using Seaborn
ax = sns.displot(netflix_df['rating']) 

plt.title('Distribution of Netflix Ratings')
plt.xlabel('Rating')
```

```python
# Creating a distribution plot for Netflix ratings using Seaborn with a title
ax = sns.displot(netflix_df['rating'])
plt.title('Distribution of Netflix Ratings')
```

```python
# Calculating and displaying the proportion of Netflix originals in the catalog
num_titles = len(netflix_df)
num_originals = netflix_df[netflix_df['type'] == 'Original'].shape[0] 

print(f"There are {num_titles} titles in the Netflix catalog.")
print(f"{num_originals / num_titles:.2%} of titles are Netflix originals.")
```

```python
# Creating a bar plot to visualize the distribution of content types in the Netflix dataset
ax = netflix_df['type'].value_counts().plot(kind='bar')
ax.set_title("Content Distribution by Type")
plt.show()
```
